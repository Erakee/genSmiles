```python
from rdkit import Chem
import nglview as nv
from openbabel import pybel

# 定义10个SMILES字符串示例
smiles_list = [
    "CCO",
    "CCC",
    "c1ccccc1",
    "C=O",
    "O=C=O",
    "CN",
    "CC(=O)O",
    "C#N",
    "C1=CC=CC=C1",
    "N#N"
]

views = []
for smiles in smiles_list:
    # 使用Open Babel将SMILES转换为mol格式（先通过Open Babel，后续再用rdkit处理）
    molblock = pybel.readstring("smi", smiles).write("mol")
    # 使用rdkit从mol格式数据构建分子对象
    mol = Chem.MolFromMolBlock(molblock)
    # 创建nglview视图
    view = nv.show_mol(mol)
    views.append(view)

# 将所有视图组合展示（水平排列）
combined_view = nv.NGLWidget()
for view in views:
    combined_view.add_component(view)

combined_view
```
